# 🟡 Medium: Implement BatchNorm

Implement **Batch Normalization** (training mode) from scratch.

$$\text{BN}(x) = \gamma \cdot \frac{x - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}} + \beta$$

where $\mu_B$ and $\sigma_B^2$ are the mean and variance computed **across the batch** (dim=0).

### Signature
```python
def my_batch_norm(x: torch.Tensor, gamma: torch.Tensor, beta: torch.Tensor, eps: float = 1e-5) -> torch.Tensor:
    # x: (N, D) — normalize each feature across all samples in the batch
```

### Rules
- Do **NOT** use `F.batch_norm`, `nn.BatchNorm1d`, etc.
- Compute mean and variance over `dim=0` with `unbiased=False`
- Must support autograd

````python
import torch
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

def my_batch_norm(x, gamma, beta, eps=1e-5):
    #先算均值和方差，这里是batchnorm是在batch维度上的，所以说需要dim = 0
    x_mean = torch.mean(x,dim = 0)
    x_var = torch.var(x,dim = 0,unbiased = False)
    temp_norm = (x-x_mean)/torch.sqrt(x_var+eps)
    return gamma*temp_norm+beta
  # Replace this
````

````python
# 🧪 Debug
x = torch.randn(8, 4)
gamma = torch.ones(4)
beta = torch.zeros(4)
out = my_batch_norm(x, gamma, beta)
print("Output shape:", out.shape)
print("Column means:", out.mean(dim=0))   # should be ~0
print("Column stds: ", out.std(dim=0))    # should be ~1
````
