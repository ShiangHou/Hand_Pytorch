# 🟢 Solution: Implement ReLU

Reference solution for the ReLU activation function.

$$\text{ReLU}(x) = \max(0, x)$$

```python
import torch
```

```python
# ✅ SOLUTION

def relu(x: torch.Tensor) -> torch.Tensor:
    #首先x>0
    return x * (x > 0).float()
```

```python
# Verify
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Output:", relu(x))
```

```python
# Run judge
from torch_judge import check
check("relu")
```

```python
#note
#首先x>0会会出现一个布尔值，我们需要吧布尔值转为浮点数，然后再
print(x>0)
print((x>0).float())
```

```python
#torch.float（）绝大多数情况下是不填的，如果想要去修改浮点数类型，可以用to来去做

x.to(dtype=torch.float32, device='mps')
```
